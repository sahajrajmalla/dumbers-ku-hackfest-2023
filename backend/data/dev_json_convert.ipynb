```python
import pandas as pd

# Read CSV file into a DataFrame
df = pd.read_csv('/users/smalla/dumbers-ku-hackfest-2023/backend/data/development_energy_data.csv')

df
```

```python
group_df = df.groupby("Development Category").agg("mean")
```

```python
group_df.columns = [i.lower().replace(" ", "_") for i in group_df.columns]
```

```python
group_df.reset_index(drop=False, inplace=True)
```

```python
# Save the DataFrame to a JSON file
group_df.to_json('development_energy.json', orient='records')
```
